```python
# imports
from pickle_ops import pickle_load
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining file paths

GA_FILE_PATH = "genetic_eval.pkl"
BO_FILE_PATH = "bayesian_eval.pkl"
```

```python
# loading final evaluation data
GA_eval = pickle_load(GA_FILE_PATH)
BO_eval = pickle_load(BO_FILE_PATH)
```

```python
# Calculating evaluation statistics
GA_min = min(GA_eval)
GA_max = max(GA_eval)
GA_avg = np.average(GA_eval)
GA_std = np.std(GA_eval)

BO_min = min(BO_eval)
BO_max = max(BO_eval)
BO_avg = np.average(BO_eval)
BO_std = np.std(BO_eval)
```

```python
# Printing results

print(f"Minimum evaluation value for GA is {GA_min}.")
print(f"Maximum evaluation value for GA is {GA_max}.")
print(f"Average evaluation value for GA is {GA_avg}.")
print(f"Standard Deviation for evaluation values for GA is {GA_std}.")
print()
print(f"Minimum evaluation value for BO is {BO_min}.")
print(f"Maximum evaluation value for BO is {BO_max}.")
print(f"Average evaluation value for BO is {BO_avg}.")
print(f"Standard Deviation for evaluation values for BO is {BO_std}.")
```

```python
# Defining training values during hyperparameter tuning
# Note: these values are hardcoded for our own results. 
#       we could not store them in pickle files due to an error.
GA_training_values = [-5580.0, -5065.0, -5065.0, -4820.0, -4820.0, -4610.0, -4610.0, -4035.0, -4035.0, -4035.0, -4035.0, -4035.0, -4035.0, -4035.0, -4035.0, -4035.0, -4805.0, -4805.0, -4805.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0, -3625.0]
BO_training_values = [-7750.0, -4950.0, -7690.0, -7850.0, -150315.0, -147395.0, -4735.0, -149200.0, -6900.0, -14260.0]
```

```python
# Graphing the values
# For GA, it is the training value with each generation.
# For BO, it is the training value with each trial.

plt.figure(figsize=(6,4))
plt.xticks(np.arange(0, 31, 2)) 
plt.xlabel('Generation', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('GA generational training scores', fontsize=14, fontweight='bold')
plt.plot([i for i in range(16)], GA_training_values[:16], '.b-')
plt.plot([i for i in range(16, 30)], GA_training_values[16:30], '.r-')
plt.legend(['5K action cap training', '10K action cap training'], loc='lower center', fontsize='small')
```

```python
plt.figure(figsize=(6,4))
plt.xticks(np.arange(0, 11, 2)) 
plt.xlabel('Trial', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Bayesian Optimization training scores', fontsize=14, fontweight='bold')
plt.plot([i for i in range(10)], BO_training_values,'.g-')   
plt.legend(['30K action cap training'], loc='lower left', fontsize='small')
```
